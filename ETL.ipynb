```python
# Dependencies
import pandas as pd
```

```python
draft_pos = "nfl_draft_position.txt"
```

```python
draft_pos_df = pd.read_csv(draft_pos)
draft_pos_df.head()
```

```python
draft_pos_df.columns
```

```python
draft_pos_df['Player'] = draft_pos_df['Player'].str.split(pat = "\\").str[0]
```

```python
draft_data = draft_pos_df[['Rk', 'Year', 'Rnd', 'Pick', 'Player', 'DrAge', 'Tm', 'From',
       'To', 'AP1', 'PB', 'St', 'CarAV', 'G', 'GS','College/Univ']]
draft_data.head()
```

```python
draft_data = draft_data.rename(columns={"Tm": "Draft_Tm"})
draft_data.head()
```

```python
salaries = pd.read_csv("nfl_salaries.txt")
salaries.head()
```

```python
salaries.columns
```

```python
salaries_df = salaries[['Player', 'Pos', 'Tm', 'Salary']]
```

```python
salaries_df['Player'] = salaries_df['Player'].str.split(pat = "\\").str[0]
salaries_df.head()
```

```python
df = pd.merge(salaries_df, draft_data, how = "inner", on = "Player")
df.head()
```
